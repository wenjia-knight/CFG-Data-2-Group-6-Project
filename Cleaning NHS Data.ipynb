```python
import pandas as pd
data = pd.read_csv("datasets/MHSDS Time_Series_data_Apr_2016_JanPrf_2023.csv")
```

```python
data_copy = data.copy()

data_copy = data_copy[['REPORTING_PERIOD_START', 'MEASURE_ID', 'MEASURE_VALUE']]
```

```python
data_copy.rename(columns={'REPORTING_PERIOD_START': 'month','MEASURE_ID': 'id', 'MEASURE_VALUE': 'new_referrals_count'}, inplace=True)
```

```python
data_copy[['month','id','new_referrals_count']] = data_copy[['month','id', 'new_referrals_count']].astype("string")
data_copy.dtypes
```

```python
data_copy = data_copy[~data_copy['new_referrals_count'].isin(['*'])]
data_copy.shape
```

```python
data_copy['new_referrals_count'] = data_copy['new_referrals_sum'].astype("float")
data_copy.dtypes
```

```python
data_copy['month'] = data_copy['month'].apply(pd.to_datetime, format= '%d/%m/%Y')
data_copy.dtypes
```

```python
data_copy.sort_values(by=['month'], ascending=True, inplace=True)
data_copy
```

```python
date_filtered_df = data_copy.loc[(data_copy['month'] >= '2018-01-01')
                                 & (data_copy['month'] < '2023-01-01')]
date_filtered_df
```

```python
id_date_filtered_df = date_filtered_df.loc[(data_copy['id'] == 'MHS32')]
id_date_filtered_df
```

```python
month_filtered_series = id_date_filtered_df.groupby('month')['new_referrals_sum'].sum()
month_filtered_df = pd.DataFrame(month_filtered_series)

month_filtered_df
```

```python
month_filtered_df.to_csv('datasets/NHS_data_cleaned.csv', encoding='utf-8')
```
