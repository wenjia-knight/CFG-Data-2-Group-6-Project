# Moods of the tracks streamed in 2018

```python
# filtering out the data in terms of year. Counting how many tracks of each mood were most streamed in 2018
df_2018 = new_track_data.query('date.dt.year == 2018')
 
df_2018 = df_2018.filter(['date', 'mood'])
 
track_moods_2018 = df_2018.groupby('mood').count()
```

```python
track_moods_2018
```

```python
fig = px.pie(track_moods_2018.reset_index(names = 'mood'), values='date', names='mood', height = 400, width = 800)
fig.update_layout(legend=dict(orientation="h", yanchor="bottom", y=-0.2, xanchor="center", x=0.5), title_text='Distribution of mood in the most streamed tracks in 2018', title_x=0.5)
fig.show()
```

The output is correct as there were a total of 1440 tracks for the  year 2018 in the dataset and the sum of all the values reported is also 1440.

# Plots of mood counts per year

```python
moods = ['happy', 'agitated', 'sad', 'calm']

for i in range(0, 4):
  df = new_track_data.copy()
  df = df[(df.mood == moods[i])]
  df['year'] = pd.DatetimeIndex(df['date']).year
  count = df.groupby(['year'])['mood'].count().plot()
  plt.xlabel('year')
  plt.ylabel('number of tracks')
  plt.title('Number of tracks of each mood per year')
  plt.legend(['happy', 'agitated', 'sad', 'calm'],  bbox_to_anchor=(0.2, 1.0), fontsize = 8.5)
  
  
```

# Plots of mood counts per year

```python
moods = ['happy', 'agitated', 'sad', 'calm']

for i in range(0, 4):
  df = new_track_data.copy()
  df = df[(df.mood == moods[i])]
  df['month_year'] = pd.to_datetime(df['date']).dt.to_period('M')
  count = df.groupby(['month_year'])['mood'].count().plot()
  plt.xlabel('month and year')
  plt.ylabel('number of tracks')
  plt.title('Number of tracks of each mood per month per year')
  plt.legend(['happy', 'agitated', 'sad', 'calm'],  fontsize = 9)
```
